```python
import numpy as np
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

```

```python
names = ['Bedroom', 'Dining', 'Forest', 'Hospital', 'Mountain', 'Office']
```

```python
mt1 = np.array([0.2827068656, 0.0004888795, 2.2038e-06, 0.00311229, 4.2396e-06, 0.0632705109])
mt2 = np.array([4.49895e-05, 0.0623895293, 9.8616e-06, 0.0002612206, 1.9985e-05, 0.0013186842])
mt3 = np.array([1.217e-06, 3.844e-07, 0.8071916352, 3.7159e-06, 0.0002138004, 2.1719e-06])
mt4 = np.array([0.0028615297, 0.0002686867, 2.2398e-06, 0.8568878536, 3.9808e-06, 0.0063317254])
mt5 = np.array([4.7996e-06, 2.6589e-06, 0.0139842389, 2.05725e-05, 0.3189772014, 1.3788e-05])
mt6 = np.array([0.0007326253, 0.0040781135, 3.7252e-06, 0.0009698673, 1.18846e-05, 0.6575974673])
```

```python
matrix = [mt1/sum(mt1),mt2/sum(mt2),mt3/sum(mt3),mt4/sum(mt4),mt5/sum(mt5),mt6/sum(mt6)]
```

```python
df_cm = pd.DataFrame(matrix, index = [i for i in names], columns = [i for i in names])
plt.figure(figsize = (15,7))
sn.heatmap(df_cm, annot=True, annot_kws={"size": 16})
plt.savefig("conf_matrix_googlenet.jpg")
```
