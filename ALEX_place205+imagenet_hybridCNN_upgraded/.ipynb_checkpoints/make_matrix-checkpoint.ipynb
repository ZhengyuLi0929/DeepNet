```python
import numpy as np
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

```

```python
names = ['Beach', 'Forest', 'Highway', 'Mountain', 'Office']
```

```python
mt1 = np.array([0.0314763689, 9.6977e-06, 1.31093e-05, 4.48978e-05, 6.9113e-06])
mt2 = np.array([6.482e-07, 0.3470846768, 1.23918e-05, 2.56536e-05, 3.425e-07])
mt3 = np.array([3.29101e-05, 0.0108592369, 0.1432023871, 0.0096132762, 7.1588e-06])
mt4 = np.array([0.0001590406, 0.0034754266, 0.0006800448, 0.0203135672, 8.251e-07])
mt5 = np.array([9.855e-07, 3.749e-07, 6.26e-07, 1.8153e-06, 0.1411490208])
```

```python
matrix = [mt1/sum(mt1),mt2/sum(mt2),mt3/sum(mt3),mt4/sum(mt4),mt5/sum(mt5)]
```

```python
df_cm = pd.DataFrame(matrix, index = [i for i in names], columns = [i for i in names])
plt.figure(figsize = (15,7))
sn.heatmap(df_cm, annot=True, annot_kws={"size": 16})
plt.savefig("goodbad_matrix_hybrid_good.png")
```
