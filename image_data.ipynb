```python
# roll
def roll_dice(num_dice = 2, num_sides = 6):
    # A fair dice can only have 4, 6, 8, 12, or 20 sides
    # here we check if the input num_sides matches
    list_of_possible_sides = [4,6,8,12,20]
    if not (num_sides in list_of_possible_sides) :
        print ("It is not possible to have a "+str(num_sides)+"-sided fair dice.")
        return
    # we do a for loop to create the return value
    return_list = []
    for roll in range(num_dice):
        # here we +1 since range() function creates values from 0 to num_sides-1
        choice = rd.choice(range(num_sides)) + 1
        return_list.append(choice)
    return_list.append(sum(return_list))
    return return_list
    

# odd number checking function
def odd_num_check(input_list):
    return_list = []
    for item in input_list:
        if type(item) != int:
            print("There is a type error in the list.")
            return
        if item % 2 == 1:
            return_list.append(item)
    return return_list
```

```python
'''
The Connection class should define a connection in a neural network. It should have a recipient (the unit on the 
receiving end of the connection), a sender (the unit on the sending end) and a weight. The __init__ method on the 
class should expect to be told who the recipient and sender will be. Extra kudos if you write the code to handle cases
where the recipient and sender are not known at the time of the Connection objects creation. You should also write 
the class so that it is possible to pass a weight into the __init__ method, but so that it defaults to a weight of 0.0
if no weight is specified. 

The Unit class should define a unit (node) in a neural network. It should have an input, an activation (both 
initialized to 0.0 in __init__) and an (initially empty) list of incoming connections (i.e., connections for which 
the Unit object in question serves as the recipient). If you like, you can also give it a list of outgoing connections
(i.e., connections for which the Unit in question serves as the sender). To be really general, write the function so 
that the unit has, not just a net input (sum of positive and negative inputs) but separate variables for inhibitory 
inputs (inputs arriving over connections with negative weights) and excitatory inputs (inputs from positive weights) 
or even separate lists for excitatory and inhibitory connections. 

The Unit class should also have at least three methods: add_connection, which connects it to another unit;  get_input,
whereby it gets inputs from other units (by default, use the dot product rule: input = sum, over all connections and
units, j, of the activation of unit j times the connection weight from j to i, where i is the Unit object itself); and
update_activation, which converts the units input to an activation. Depending on the activation function you use, the
Unit class may also need a threshold (defined in __init__). 
'''

import numpy as np

class connection():
    def __init__(self, sender, recipient, weight = 0.0):
        self.sender = sender
        self.recipient = recipient
        self.weight = weight
        
class Unit():
    # I set inhibitory and excitatory inputs to be two separate lists
    # And input is just the net input
    def __init__(self, input_ = 0.0, inh_input = [], exc_input = [], activation = 0.0, connections = []):
        self.input = input_
        self.activation = activation
        self.connections = connections
        self.inh = inh_input
        self.exc = exc_input
    
    # we use the connection class to create a connection
    def add_connection(self, node, weight):
        self.connections.append(connection(node, self, weight))
    
    # implement sigmoid activation function
    def sigmoid_activation(self, x):
        return 1/(1+np.exp(-x))
    
    # we get input by looping connections
    def get_input(self):
        # check if they were inhibitory or excitatory
        for item in self.connections:
            if item.weight < 0:
                self.inh.append(item.sender.input * item.weight)
            else:
                self.exc.append(item.sender.input * item.weight)
        # adding them up, and activate it.
        self.input = self.sigmoid_activation(sum(self.inh) + sum(self.exc))
```
