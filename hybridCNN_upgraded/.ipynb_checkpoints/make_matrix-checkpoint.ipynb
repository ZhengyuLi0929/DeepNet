```python
import numpy as np
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

```

```python
names = ['Bedroom', 'Dining', 'Forest', 'Hospital', 'Mountain', 'Office']
```

```python
mt1 = np.array([0.0577769997, 3.54898e-05, 2.83e-08, 0.0014804654, 3.07e-07, 0.0005327469])
mt2 = np.array([2.8768e-06, 0.0006482014, 7.41e-08, 9.0803e-06, 1.0547e-06, 0.000128259])
mt3 = np.array([7e-08, 2.445e-07, 0.3470846768, 2.978e-07, 2.56536e-05, 3.425e-07])
mt4 = np.array([0.0001488816, 5.44683e-05, 1.649e-07, 0.3075043187, 1.2207e-06, 0.0001952693])
mt5 = np.array([4.998e-07, 3.507e-07, 0.0046060997, 8.652e-07, 0.0267266448, 1.069e-06])
mt6 = np.array([6.07187e-05, 0.0022882838, 3.749e-07, 0.0011982263, 1.8153e-06, 0.1411490208])
```

```python
matrix = [mt1/sum(mt1),mt2/sum(mt2),mt3/sum(mt3),mt4/sum(mt4),mt5/sum(mt5),mt6/sum(mt6)]
```

```python
df_cm = pd.DataFrame(matrix, index = [i for i in names], columns = [i for i in names])
plt.figure(figsize = (15,7))
sn.heatmap(df_cm, annot=True, annot_kws={"size": 16})
plt.savefig("conf_matrix.png")
```
