```python
import numpy as np
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

```

```python
names = ['Bedroom', 'Dining', 'Forest', 'Hospital', 'Mountain', 'Office']
```

```python
mt1 = np.array([0.3302040056, 0.0026586302, 7.298e-07, 0.006738205, 2.7283e-06, 0.0313056111])
mt2 = np.array([0.000199937, 0.0426168813, 4.3019e-06, 0.0004327681, 2.58649e-05, 0.0031319492])
mt3 = np.array([2.119e-07, 2.232e-07, 0.7511114801, 1.4967e-06, 0.0003470396, 1.0375e-06])
mt4 = np.array([0.0110558004, 0.0009039788, 2.4108e-06, 0.8755645496, 3.7239e-06, 0.003471136])
mt5 = np.array([1.5506e-06, 1.2048e-06, 0.0105894786, 5.3478e-06, 0.3320629702, 5.3235e-06])
mt6 = np.array([0.0005937072, 0.021333471, 2.5805e-06, 0.0059224993, 1.22642e-05, 0.5279001569])
```

```python
matrix = [mt1/sum(mt1),mt2/sum(mt2),mt3/sum(mt3),mt4/sum(mt4),mt5/sum(mt5),mt6/sum(mt6)]
```

```python
df_cm = pd.DataFrame(matrix, index = [i for i in names], columns = [i for i in names])
plt.figure(figsize = (15,7))
sn.heatmap(df_cm, annot=True, annot_kws={"size": 16})
plt.savefig("conf_matrix.png")
```
