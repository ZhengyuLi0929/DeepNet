```python
# First, we import the network structure we implemented, with some adjustifications.
# import numpy module to easily implement sigmoid activation function
import numpy as np

# Unit class
class Unit():
    # I set inhibitory and excitatory inputs to be two separate lists
    # And input is just the net input
    def __init__(self, input_ = 0.0, activation = 0.0, connections = []):
        self.input = input_
        self.activation = activation
        self.connections = connections
        
    # we get input by looping connections
    def get_input(self):
        value = []
        for item in self.connections:
            value.append(item[0].activation * item[1])
        self.input += sum(value)
        
# here we have the two activation functions with part 1 1) and 2)
def activation1(n):
    return n - 0.5
def activation2(n):
    if n < 0.75:
        return 0
    elif n > 1:
        return 1
    else:
        return n
```

```python
# number of iterations
iteration = 10

# Activation case (a)
nodes = []
A = Unit(activation = 1.0)
B = Unit(activation = 1.0)
C = Unit()
nodes.append(A)
nodes.append(B)
nodes.append(C)
A.connections = [[B, 1.0], [C, 1.0]]
B.connections = [[A, 1.0], [C, 1.0]]
C.connections = [[A, 1.0], [B, 1.0]]

# begin simulation for 1)
print("CASE 1) a)")
for i in range(iteration):
    for node in nodes:
        node.get_input()
    for node in nodes:
        node.activation = activation1(node.input)
    print("Round",i+1,"activation:",A.activation, B.activation, C.activation)
    
# begin simulation for 2)
print("CASE 2) a)")
for i in range(iteration):
    for node in nodes:
        node.get_input()
    for node in nodes:
        node.activation = activation2(node.input)
    print("Round",i+1,"activation:",A.activation, B.activation, C.activation)
```

```python
# PART 2
# Kinda new Structure

# Unit class
class Unit():
    # And input is just the net input
    def __init__(self, input_ = 0.0, activation = 0.0, connections = [], inh = 0.0, exc = 0.0):
        self.input = input_
        self.activation = activation
        self.connections = connections
        self.inh = inh
        self.exc = exc
        
    # we get input by looping connections
    def get_input(self):
        iv = []
        ev = []
        for item in self.connections:
            if item[0].activation < 0:
                iv.append(-item[0].activation * item[1])
            else:
                ev.append(item[0].activation * item[1])
        self.input += sum(ev) + sum(iv)
        
# activation functions for part 2, solve the math
# a) ∆ai = γ(1-ai)ni, where γ (gamma) = 0.5
def activation_A(a, n):
    return 0.5 * (1-a)*n

#b) ∆ai = γ(1-ai)ni - δai, where γ = 0.5 and δ (delta) = 0.1
def activation_B(a,n):
    return 0.5 * (1-a) * n - 0.1 * a

#c) ∆ai = γ[(1-ai)ei + (1 + ai)ii] - δai, where γ = 0.5 and δ = 0.1 (assume ii is negative)
def activation_C(a, n, exc, inh):
    return 0.5 *((1-a)*exc + (1+a)*inh) - 0.1 * a
```

```python
# a1) Run rule (a) for 10 iterations with ni = 1.5 on each iteration

nodes = []
A = Unit()
B = Unit()
C = Unit()
nodes.append(A)
nodes.append(B)
nodes.append(C)
A.connections = [[B, 1.0], [C, 1.0]]
B.connections = [[A, 1.0], [C, 1.0]]
C.connections = [[A, 1.0], [B, 1.0]]

# begin simulation for 1)
for i in range(iteration):
    for node in nodes:
        node.get_input()
    for node in nodes:
        node.activation += activation_A(node.input, 1.5)
    print("Round",i+1,"activation:",A.activation, B.activation, C.activation)
```
