```python
import numpy as np
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

```

```python
names = ['Bedroom', 'Dining', 'Forest', 'Hospital', 'Mountain', 'Office']
```

```python
mt1 = np.array([0.2877214841, 1.64042e-05, 2.153e-07, 0.0004533898, 2.4396e-06, 0.1004566619])
mt2 = np.array([1.89624e-05, 0.0683714663, 5.9773e-06, 4.32006e-05, 2.17036e-05, 0.0003743109])
mt3 = np.array([6.8e-09, 1.05e-08, 0.8347187762, 9.72e-08, 0.0001174185, 1.178e-07])
mt4 = np.array([0.0002961912, 0.0002790905, 2.482e-07, 0.8488076436, 4.167e-07, 0.0382160434])
mt5 = np.array([1.4377e-06, 9.139e-07, 0.0056651683, 4.4983e-06, 0.3637920283, 7.1358e-06])
mt6 = np.array([8.28128e-05, 0.0009945967, 4.841e-07, 0.0003593523, 4.201e-07, 0.6850329464])
```

```python
matrix = [mt1/sum(mt1),mt2/sum(mt2),mt3/sum(mt3),mt4/sum(mt4),mt5/sum(mt5),mt6/sum(mt6)]
```

```python
df_cm = pd.DataFrame(matrix, index = [i for i in names], columns = [i for i in names])
plt.figure(figsize = (15,7))
sn.heatmap(df_cm, annot=True, annot_kws={"size": 16})
plt.savefig("conf_matrix.jpg")
```
