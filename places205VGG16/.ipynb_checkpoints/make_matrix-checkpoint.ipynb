```python
import numpy as np
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

```

```python
names = ['Beach', 'Forest', 'Highway', 'Mountain', 'Office']
```

```python
mt1 = np.array([0.2690883749, 0.0015449923, 1.73983e-05, 0.0155561759, 1.3721e-05])
mt2 = np.array([2.4753e-06, 0.8347187762, 2.37047e-05, 0.0001174185, 1.178e-07])
mt3 = np.array([0.0006746956, 0.1756446699, 0.6238100706, 0.020135045, 0.0020280229])
mt4 = np.array([0.0001356725, 0.0043771316, 0.0006035486, 0.2745699407, 5.6991e-06])
mt5 = np.array([4.207e-07, 4.841e-07, 2.887e-07, 4.201e-07, 0.6850329464])
```

```python
matrix = [mt1/sum(mt1),mt2/sum(mt2),mt3/sum(mt3),mt4/sum(mt4),mt5/sum(mt5)]
```

```python
df_cm = pd.DataFrame(matrix, index = [i for i in names], columns = [i for i in names])
plt.figure(figsize = (15,7))
sn.heatmap(df_cm, annot=True, annot_kws={"size": 16})
plt.savefig("goodbad_matrix_place_good.jpg")
```
