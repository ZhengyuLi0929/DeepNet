```python
import os
for directory in directories:
    newdir = os.path.join(directory, 'flipped')
    if not os.path.exists(newdir):
        os.mkdir(newdir)
    for filename in os.listdir(directory):
        if filename.endswith(".jpeg") or filename.endswith(".jpg"):
            imgpath = os.path.join(directory, filename)
            im = Image.open(imgpath)
            flipped = ImageOps.flip(im)
            newimgname = "flipped_" + filename
            newpath = os.path.join(newdir, newimgname)
            flipped.save(newpath)
```

```python
for directory in directories:
    for filename in os.listdir(directory):
        if filename.endswith(".jpeg") or filename.endswith(".jpg"):
            if filename[0] == 'f' and filename[1] == 'l':
                imgpath = os.path.join(directory, filename)
                os.remove(imgpath)
```
